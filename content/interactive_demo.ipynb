## 1. Dynamic Economic Scenarios
Use the slider below to simulate an interest rate adjustment and see the impact on a hypothetical model.

```python
import ipywidgets as widgets
from IPython.display import display

slider = widgets.FloatSlider(
    value=4.5, 
    min=0, 
    max=10, 
    step=0.25, 
    description='Interest Rate %:',
    style={'description_width': 'initial'}
)
display(slider)
```

## 2. Geospatial Analysis
With `ipyleaflet`, we can visualize global financial cycles or regional data.

```python
from ipyleaflet import Map, Marker

center = [55.6761, 12.5683] # Copenhagen
m = Map(center=center, zoom=10)
marker = Marker(location=center, draggable=False)
m.add_layer(marker)
m
```
